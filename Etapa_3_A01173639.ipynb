```python

from funpymodeling.exploratory import freq_tbl 
```

```python
#Carga desde un archivo .xlsx sin indice
df_eua= pd.read_csv('data_clean_california_eua.csv') 
df_esp= pd.read_csv('data_clean_girona_españa.csv') 
df_mex= pd.read_csv('data_clean_df_mexico.csv') 
```

```python
tf_df_eua=freq_tbl(df_eua)
```

```python
#verificamos nulos de los frames
df_mex=df_mex.fillna(method="ffill") 
df_mex.isnull().sum().sum()
df_esp=df_esp.fillna(method="ffill") 
df_esp.isnull().sum().sum()

nulos_eua=df_eua.isnull().sum().sum()
nulos_esp=df_esp.isnull().sum().sum()
nulos_mex=df_mex.isnull().sum().sum()

print(nulos_eua)
print(nulos_esp)
print(nulos_mex)
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1_eua= freq_tbl(df_eua['host_identity_verified'])
table1_eua= table1_eua[table1_eua['frequency']>1]
table1_eua= table1_eua.set_index('host_identity_verified')

table1_esp=freq_tbl(df_esp['host_identity_verified'])
table1_esp= table1_esp[table1_esp['frequency']>1]
table1_esp= table1_esp.set_index('host_identity_verified')

table1_mex=freq_tbl(df_mex['host_identity_verified'])
table1_mex= table1_mex[table1_mex['frequency']>1]
table1_mex= table1_mex.set_index('host_identity_verified')

print("variable 1 = EUA-host_identity_verified")
print(table1_eua)
print("\n")
print("variable 1 = ESP-host_identity_verified")
print(table1_esp)
print("\n")
print("variable 1 = MEX-host_identity_verified")
print(table1_mex)
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
table2_eua= freq_tbl(df_eua['property_type'])
table2_eua=table2_eua[table2_eua['frequency']>10]
table2_eua= table2_eua.set_index('property_type')

table2_esp=freq_tbl(df_esp['property_type'])
table2_esp=table2_esp[table2_esp['frequency']>10]
table2_esp= table2_esp.set_index('property_type')

table2_mex=freq_tbl(df_mex['property_type'])
table2_mex= table2_mex[table2_mex['frequency']>10]
table2_mex= table2_mex.set_index('property_type')

print("variable 2 = EUA-property_type")
print(table2_eua)
print("\n")
print("variable 2 = ESP-property_type")
print(table2_esp)
print("\n")
print("variable 2 = MEX-property_type")
print(table2_mex)
```
